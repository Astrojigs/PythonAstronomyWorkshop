### **Part 1: Writing Python Functions**
#### **Learning Objectives**:
- Understand how to define and call functions in Python.
- Learn about parameters, return values, and default arguments.
- Recognize the importance of modular programming.

#### **Key Concepts**:
1. **Definition and Syntax**:
   - Use the `def` keyword.
   - Syntax: 
     ```python
     def function_name(parameters):
         """Docstring describing the function."""
         body
         return value
     ```

2. **Parameters and Arguments**:
   - Positional and keyword arguments.
   - Default arguments for flexibility.

3. **Scope of Variables**:
   - Local vs. global variables.

#### **Examples**:
1. **Simple Function**:
   ```python
   def greet(name):
       return f"Hello, {name}!"
   
   print(greet("Wilson College"))
   ```

2. **Function with Default Arguments**:
   ```python
   def calculate_area(length, width=10):
       return length * width
   
   print("Area:", calculate_area(5))
   ```

3. **Factorial Using Recursion**:
   ```python
   def factorial(n):
       if n == 0:
           return 1
       return n * factorial(n - 1)
   
   print("Factorial of 5:", factorial(5))
   ```

### **Part 2: Curve Fitting Using SciPy**
#### **Learning Objectives**:
- Understand what curve fitting is and why it is useful.
- Learn how to use `scipy.optimize.curve_fit`.
- Fit a mathematical model to data.

#### **Key Concepts**:
1. **What is Curve Fitting?**:
   - Process of finding a curve that best represents the relationship between variables.

2. **Mathematical Models**:
   - Linear, polynomial, or sinusoidal models.
   - Example: `y = a * x + b` (linear), `y = a * sin(b * x) + c` (sinusoidal).

3. **Using `curve_fit`**:
   - Requires a function model, data (x and y), and initial guesses.

#### **Examples**:
1. **Fitting a Linear Model**:
   ```python
   import numpy as np
   from scipy.optimize import curve_fit
   import matplotlib.pyplot as plt

   # Define a linear model
   def linear_model(x, a, b):
       return a * x + b

   # Generate synthetic data
   x_data = np.linspace(0, 10, 50)
   y_data = 3 * x_data + 2 + np.random.normal(0, 1, size=len(x_data))

   # Perform curve fitting
   params, covariance = curve_fit(linear_model, x_data, y_data)
   a, b = params

   # Plot the data and the fit
   plt.scatter(x_data, y_data, label="Data")
   plt.plot(x_data, linear_model(x_data, a, b), color="red", label="Fit")
   plt.legend()
   plt.title("Linear Curve Fitting")
   plt.xlabel("x")
   plt.ylabel("y")
   plt.show()
   ```

2. **Fitting a Sinusoidal Model**:
   ```python
   # Define sinusoidal model
   def sinusoidal_model(x, a, b, c):
       return a * np.sin(b * x) + c

   # Generate synthetic data
   x_data = np.linspace(0, 2 * np.pi, 100)
   y_data = 2 * np.sin(1.5 * x_data) + 1 + np.random.normal(0, 0.2, len(x_data))

   # Perform curve fitting
   params, covariance = curve_fit(sinusoidal_model, x_data, y_data, p0=[2, 1.5, 1])
   a, b, c = params

   # Plot the data and the fit
   plt.scatter(x_data, y_data, label="Data")
   plt.plot(x_data, sinusoidal_model(x_data, a, b, c), color="red", label="Fit")
   plt.legend()
   plt.title("Sinusoidal Curve Fitting")
   plt.xlabel("x")
   plt.ylabel("y")
   plt.show()
   ```

````python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Define sinusoidal model
def sinusoidal_model(x, a, b, c):
    return a * np.sin(b * x) + c

# Generate synthetic data
x_data = np.linspace(0, 2 * np.pi, 100)
y_data = 2 * np.sin(1.5 * x_data) + 1 + np.random.normal(0, 0.2, len(x_data))

# Perform curve fitting
params, covariance = curve_fit(sinusoidal_model, x_data, y_data, p0=[2, 1.5, 1])
a, b, c = params

# Calculate fitted values and residuals
y_fitted = sinusoidal_model(x_data, a, b, c)
residuals = y_data - y_fitted

# Plot residuals
plt.figure(figsize=(8, 5))
plt.scatter(x_data, residuals, label="Residuals", color="orange")
plt.axhline(0, color='blue', linestyle='--', linewidth=1, label="Zero Line")
plt.title("Residuals of Sinusoidal Fit")
plt.xlabel("x")
plt.ylabel("Residuals (Observed - Fitted)")
plt.legend()
plt.show()
````

Here are a few bullet points to explain the residuals chart:

- **Definition of Residuals**: Residuals are the differences between the observed data points and the predicted values from the fitted model. They are calculated as $ \text{Residual} = \text{Observed Value} - \text{Fitted Value}$.

- **Purpose of the Chart**:
  - The residuals chart helps assess the quality of the fit.
  - A good fit will have residuals scattered randomly around the zero line, indicating no systematic errors in the model.

- **Key Observations from the Chart**:
  - **Random Distribution**: The residuals in this chart appear mostly scattered around the zero line, suggesting that the sinusoidal model captures the overall trend well.
  - **No Clear Pattern**: If residuals show a trend or pattern (e.g., a wave-like structure), it indicates the model may not fully capture the underlying data behavior.
  - **Magnitude of Residuals**: The spread of residuals gives an idea of the model's accuracy. Smaller residuals indicate a better fit.

- **Interpretation**: The chart shows that the sinusoidal model provides a reasonable fit to the data, as the residuals are small and do not exhibit systematic patterns. However, slight deviations can indicate areas where the model might be improved.
